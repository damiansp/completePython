```python
im = imread(f'{IMG}/Ishihara.png')
im_g = color.rgb2gray(im)

plt.subplot(121)
plt.imshow(im)
plt.axis('off')

plt.subplot(122)
plt.imshow(im_g, cmap='gray')
plt.axis('off');
```

```python
im = imread(f'{IMG}/parrot.png')
im_hsv = color.rgb2hsv(im)

plt.gray()
plt.figure(figsize=(10, 8))

plt.subplot(221)
plt.imshow(im_hsv[..., 0])
plt.title('h', size=20)
plt.axis('off')

plt.subplot(222)
plt.imshow(im_hsv[..., 1])
plt.title('s', size=20)
plt.axis('off')

plt.subplot(223)
plt.imshow(im_hsv[..., 2])
plt.title('v', size=20)
plt.axis('off')

plt.subplot(224)
plt.axis('off');
```
