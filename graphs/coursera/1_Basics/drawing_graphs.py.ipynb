```python
import networkx as nx
import numpy as np
import pygraphviz as pgv
import nxpd
from nxpd import draw, nxpdParams

nxpdParams['show'] = 'ipynb'
```

```python
G = nx.DiGraph()
G.add_edge('a', 'b')
G.add_edge('b', 'c')
G.add_edge('c', 'd')
G.add_edge('d', 'e')
G.add_edge('e', 'c')
G.add_edge('a', 'd')
```

```python
draw(G, layout='circo')
```

```python
draw(G, layout='dot')
```

```python
draw(G, layout='neato')
```

```python
G = nx.DiGraph()
nodes = list('abcdefghij')
n_edges = 20
for edge in range(n_edges):
    e1, e2 = np.random.choice(nodes, 2)
    G.add_edge(e1, e2)
```

```python
draw(G, layout='circo')
```

```python
draw(G, layout='dot')
```

```python
draw(G, layout='neato')
```
