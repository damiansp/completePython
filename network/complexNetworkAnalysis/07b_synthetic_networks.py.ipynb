# Slicing Weighted Networks
(Removing edges with low weights)

```python
G = nx.erdos_renyi_graph(30, 0.2)
```

```python
draw(G, layout='circo')
```

```python
def slice_network(G, threshold, copy=True):
    '''Remove all edges of weight < <threshold>'''
    F = G.copy() if copy else G
    F.remove_edges_from((n1, n2) for n1, n2, w in F.edges(data='weight')
                        if w < threshold)
```
