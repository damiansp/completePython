```python
import networkx as nx
import pygraphviz as pgv
from nxpd import draw, nxpdParams

nxpdParams['show'] = 'ipynb'
```

```python
G = nx.Graph()
G.add_edges_from([
    (1, 'b'), (1, 'c'), (1, 'd'), (2, 'a'), (2, 'c'), (2, 'e'), (3, 'b'),
    (3, 'c'), (3, 'd'), (4, 'a'), (4, 'e'), (5, 'a'), (5, 'e')])
```

```python
if nx.bipartite.is_bipartite(G):
    left, right = nx.bipartite.sets(G)
    for v in left:
        G.node[v]['color'] = 'blue'
    for v in right:
        G.node[v]['color'] = 'red'
else:
    print('Graph is not bipartite')
```

```python
draw(G, layout='circo')
```

```python
M = nx.max_weight_matching(G)
print(type(M))
M
```

```python
for v1, v2 in M:
    G[v1][v2]['color'] = 'green'

draw(G, layout='circo')
```
