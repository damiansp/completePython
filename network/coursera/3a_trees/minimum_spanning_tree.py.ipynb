```python
import networkx as nx
import numpy as np
import pygraphviz as pgv
from nxpd import draw, nxpdParams

nxpdParams['show'] = 'ipynb'
```

```python
G = nx.Graph()
weights = [1, 2, 3, 4, 5]
G.add_edges_from([
    ('a', 'b', {'weight': np.random.choice(weights), 'label': 2}), 
    ('a', 'c', {'weight': np.random.choice(weights), 'label': 3}), 
    ('a', 'd', {'weight': np.random.choice(weights), 'label': 1}), 
    ('a', 'e', {'weight': np.random.choice(weights), 'label': 3}), 
    ('b', 'c', {'weight': np.random.choice(weights), 'label': 4}), 
    ('c', 'd', {'weight': np.random.choice(weights), 'label': 5}), 
    ('d', 'e', {'weight': np.random.choice(weights), 'label': 4}), 
    ('e', 'a', {'weight': np.random.choice(weights), 'label': 1})])
draw(G, layout='circo')
```

```python
T = nx.minimum_spanning_tree(G)
for e in T.edges():
    G[e[0]][e[1]]['color'] = 'red'
draw(G, layout='circo')
```
