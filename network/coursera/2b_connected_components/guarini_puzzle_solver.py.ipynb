```python
import itertools as it

import networkx as nx
```

```python
G = nx.Graph()

# Notation
# Cells numbered as
# 0 1 2
# 3   4
# 5 6 7
# Hence W*W**B*B represents white knights at 0 and 2 and black ones at 5 & 
# 7

# all quadruples using digits 0-8 once each = All possible configurations
# of 2W and 2B knights on edges of a 3x3 chess board (middle cell is 
# unreachable, so omitted)
for wb_indices in it.permutations(range(8), 4):
    config = ['*'] * 8
    config[wb_indices[0]] = 'W'
    config[wb_indices[1]] = 'W'
    config[wb_indices[2]] = 'B'
    config[wb_indices[3]] = 'B'
    G.add_node(''.join(config))
```

```python
moves = [[] for _ in range(8)]
moves[0] = [4, 6] # from 0, 4 & 6 are reachable
moves[1] = [5, 7]
moves[2] = [3, 6]
moves[3] = [2, 7]
moves[4] = [0, 5]
moves[5] = [1, 4]
moves[6] = [0, 2]
moves[7] = [1, 3]
```

```python
for node in G.nodes():
    config = [c for c in node]
    for i in range(8):
        if config[i] != '*':
            for new_pos in moves[i]        :
                if config[new_pos] != '*':
                    continue # position already occupied
                new_config = list(config)
                new_config[i] = '*'
                new_config[new_pos] = config[i]
                if not G.has_edge(''.join(config), ''.join(new_config)):
                    G.add_edge(''.join(config), ''.join(new_config))
```

```python
print(f'n nodes: {nx.number_of_nodes(G)}')
print(f'n edges: {nx.number_of_edges(G)}')
print(f'n connected components: {nx.number_connected_components(G)}')
```

```python
# One component is the set of arangements where B & W are interchanged the
# other is where they are not: neither arrangement is reachable from the 
# other.  Verify:
assert 'W*B**W*B' in nx.node_connected_component(G, 'W*W**B*B')
assert 'B*B**W*W' in nx.node_connected_component(G, 'W*W**B*B')
assert 'W*B**B*B' not in nx.node_connected_component(G, 'W*W**B*B')
```

```python
print(' -> '.join(nx.shortest_path(G, 'W*W**B*B', 'B*B**W*W')))
```
