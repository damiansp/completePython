```python
import networkx as nx
#import pygraphviz as pgv
from nxpd import draw, nxpdParams

nxpdParams['show'] = 'ipynb'
```

```python
G = nx.Graph()
G.add_edges_from([
    ('a', 'b'), ('t', 'c'), ('b', 'x'), ('c', 'q'), ('q', 't')])
draw(G, layout='circo')
```

```python
for i, cc in enumerate(nx.connected_components(G)):
    print('%d: %s' % (i, cc))
```
