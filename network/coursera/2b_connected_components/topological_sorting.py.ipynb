```python
import networkx as nx
import pygraphviz as pgv
from nxpd import draw, nxpdParams

nxpdParams['show'] = 'ipynb'
```

```python
G = nx.DiGraph()
G.add_edges_from([
    ('a', 'b'), ('b', 'c'), ('b', 'd'), ('d', 'c'), ('a', 'd')])
draw(G, layout='circo')
```

```python
if nx.is_directed_acyclic_graph(G):
    print('Topological ordering of nodes:', nx.topological_sort(G))
else:
    print('G contains cycle(s) and cannot be topologically sortede')
```

```python
draw(G, layout='dot')
```
