```python
import polars as pl
```

```python
df = pl.DataFrame({'a': [1, 2, 3], 'b': [10, 20, 30]})
df
```

```python
(df
 .select(
     pl.fold(
         acc=pl.lit(0), f=lambda acc, x: acc + x, exprs=pl.col('*'))
     .alias('sum')))
```

```python
df = pl.DataFrame({'a': [1, 2, 3], 'b': [0, 1, 2]})
df
```

```python
(df
 .filter(
     pl.fold(
         acc=pl.lit(True),
         f=lambda acc,
         x: acc & x,
         exprs=pl.col('*') > 1)))
```

```python
df = pl.DataFrame({'a': list('abc'), 'b': [1, 2, 3]})
df
```

```python
df.select([pl.concat_str(['a', 'b'])])
```
