```python
import polars as pl
```

```python
df = pl.read_csv('https://j.mp/iriscsv')
df.head()
```

```python
(df
 .filter(pl.col('sepal_length') > 5)
 .groupby('species', maintain_order=True)
 .agg(pl.all().sum()))
```

```python
query_plan = (
    pl
    .read_csv('https://j.mp/iriscsv')
    .lazy()
    .filter(pl.col('sepal_length') > 5)
    .groupby('species', maintain_order=True)
    .agg(pl.all().sum()))
query_plan
```

```python
query_plan.collect()
```
