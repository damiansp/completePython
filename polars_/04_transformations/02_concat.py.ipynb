```python
import polars as pl
```

```python
v1 = pl.DataFrame({'a': [1], 'b': [3]})
v2 = pl.DataFrame({'a': [2], 'b': [4]})
pl.concat([v1, v2], how='vertical')
```

```python
v3 = pl.DataFrame({'x': [5], 'y': [6]})
pl.concat([v1, v3], how='horizontal')
```

```python
pl.concat([v1, v3], how='diagonal')
```
