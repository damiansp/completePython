```python
import polars as pl
```

```python
df_customers = pl.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['Amira', 'Beetlegeuse', 'Charybdis']})
df_customers
```

```python
df_orders = pl.DataFrame({
    'order_id': ['a', 'b', 'c'],
    'customer_id': [1, 2, 2],
    'amount': [100, 200, 300]})
```

```python
df_customers.join(df_orders, on='customer_id', how='inner')
```

```python
df_customers.join(df_orders, on='customer_id', how='left')
```
