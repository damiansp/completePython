```python
import polars as pl
```

```python
df = pl.DataFrame({
    'foo': list('AABBC'),
    'N': [1, 2, 2, 4, 2],
    'bar': list('klmno')})
df
```

```python
# eager
df.pivot(index='foo', columns='bar', values='N', aggregate_function='first')
```

```python
# lazy
q = (
    df
    .lazy()
    .collect()
    .pivot(index='foo', columns='bar', values='N', aggregate_function='first')
    .lazy())
q.collect()
```
