```python
sym_x = sympy.Symbol('x')
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
```

```python
def sin_exp(x, n):
    '''Eval nth order Taylor series expansion of sin(x)'''
    return sympy.lambdify(sym_x, 
                          sympy.sin(sym_x).series(n=n + 1).removeO(), 
                          'numpy')(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), linewidth=4, color='red', label='exact')
colors = ['blue', 'black']
ltys = [':', '-.', '--']
for i, n in enumerate(range(1, 12, 2)):
    ax.plot(x, 
            sin_exp(x, n), 
            color=colors[i // 3], 
            linestyle=ltys[i % 3], 
            linewidth=3, 
            label=f'O({n + 1}) appx.')
ax.set_ylim(-1.1, 1.1)
ax.set_xlim(-1.5 * np.pi, 1.5 * np.pi)
ax.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
fig.subplots_adjust(right=0.75)
```

```python
def f(x):
    return 1/(1 + x**2) + 0.1/(1 + ((3 - x)/0.1)**2)
```

```python
def plot_and_format_axes(ax, x, f, fontsize):
    ax.plot(x, f(x), lw=2)
    ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
    ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
    ax.set_xlabel(r'$x$', fontsize=fontsize)
    ax.set_ylabel(r'$f(x)$', fontsize=fontsize)
```

```python
fig = plt.figure(figsize=(8, 4))    

ax = fig.add_axes([0.1, 0.15, 0.8, 0.8], facecolor='#d3e4f5')
x = np.linspace(-4, 14, 1000)
plot_and_format_axes(ax, x, f, 18)

# inset
x0, x1 = 2.5, 3.5
ax.axvline(x0, ymax=0.3, color='grey', linestyle=':')
ax.axvline(x1, ymax=0.3, color='grey', linestyle=':')
ax_insert = fig.add_axes([0.5, 0.5, 0.38, 0.42], facecolor='#f5e4d3')
x = np.linspace(x0, x1, 1000)
plot_and_format_axes(ax_insert, x, f, 14)
```

```python
fig, axes = plt.subplots(
    1, 3, figsize=(14, 6), subplot_kw={'projection': '3d'})

def title_and_labels(ax, title):
    ax.set_title(title)
    ax.set_xlabel('$x$', fontsize=16)
    ax.set_ylabel('$y$', fontsize=16)
    ax.set_zlabel('$z$', fontsize=16)
    
x = y = np.linspace(-3, 3, 74)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(4 * R) / R
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())
p = axes[0].plot_surface(
    X, 
    Y, 
    Z, 
    rstride=1, 
    cstride=1, 
    linewidth=0, 
    antialiased=False, 
    norm=norm, 
    cmap=mpl.cm.Blues)
cb = fig.colorbar(p, ax=axes[0], shrink=0.6)
title_and_labels(axes[0], 'plot_surface')

p = axes[1].plot_wireframe(X, Y, Z, rstride=2, cstride=2, color='darkgrey')
title_and_labels(axes[1], 'plot_wireframe')

cset = axes[2].contour(
    X, Y, Z, zdir='z', offset=0, norm=norm, cmap=mpl.cm.Blues)
cset = axes[2].contour(
    X, Y, Z, zdir='y', offset=3, norm=norm, cmap=mpl.cm.Blues)
title_and_labels(axes[2], 'contour')
```
