```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import sympy
from mpl_toolkits.mplot3d.axes3d import Axes3D

%matplotlib inline
```

```python
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10*x
y3 = 6*x + 10
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='y(x)')
ax.plot(x, y2, color='purple', label="y'(x)")
ax.plot(x, y3, color='red', label='y"(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend();
```

```python
fig = plt.figure(figsize=(8, 4), facecolor='#333333')
left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax = fig.add_axes((left, bottom, width, height), facecolor='black')
x = np.linspace(-2, 2, 1000)
y1 = np.cos(40 * x)
y2 = np.exp(-x ** 2)
ax.plot(x, y1 * y2)
ax.plot(x, y2, 'y')
ax.plot(x, -y2, 'y')
ax.set_xlabel('x')
ax.set_ylabel('y');
```
