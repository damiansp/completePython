```python
x = np.linspace(-5, 5, 5)
y = np.ones_like(x)
```

```python
def axes_settings(fig, ax, title, ymax):
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(0, ymax + 1)
    ax.set_title(title)
```

```python
fix, axes = plt.subplots(1, 4, figsize=(16, 3))

# Line widths
lwds = [0.5, 1., 2., 4.]
for n, lwd in enumerate(lwds):
    axes[0].plot(x, y + n, color='blue', linewidth=lwd)
axes_settings(fig, axes[0], 'linewidth', len(lwds))

# Line style
ltys = ['-', '-.', ':']
for n, lty in enumerate(ltys):
    axes[1].plot(x, y + n, color='blue', lw=2, linestyle=lty)
    
# custom line
line, = axes[1].plot(x, y + 3, color='blue', lw=2)
length1, gap1, length2, gap2 = 10, 7, 20, 7
line.set_dashes([length1, gap1, length2, gap2])
axes_settings(fig, axes[1], 'linetypes', len(ltys) + 1)

# Marker types
markers = ['+', 'o', '*', 's', '.', '1', '2', '3', '4']
for n, marker in enumerate(markers):
    axes[2].plot(x, y + n, color='blue', lw=2, ls='None', marker=marker)
axes_settings(fig, axes[2], 'markers', len(markers))

# Marker size and color
marker_sizecolors = [(4, 'white'), (8, 'red'), (12, 'yellow'), 
                     (16, 'lightgreen')]
for n, (marker_size, marker_color) in enumerate(marker_sizecolors):
    axes[3].plot(x, 
                 y + n, 
                 color='blue', 
                 lw=1, 
                 ls='-', 
                 marker='o', 
                 markersize=marker_size, 
                 markerfacecolor=marker_color, 
                 markeredgewidth=2)
axes_settings(fig, axes[3], 'marker size/color', len(marker_sizecolors))
```

```python
sym_x = sympy.Symbol('x')
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
```

```python
def sin_exp(x, n):
    '''Eval nth order Taylor series expansion of sin(x)'''
    return sympy.lambdify(sym_x, 
                          sympy.sin(sym_x).series(n=n + 1).removeO(), 
                          'numpy')(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), linewidth=4, color='red', label='exact')
colors = ['blue', 'black']
ltys = [':', '-.', '--']
for i, n in enumerate(range(1, 12, 2)):
    ax.plot(x, 
            sin_exp(x, n), 
            color=colors[i // 3], 
            linestyle=ltys[i % 3], 
            linewidth=3, 
            label=f'O({n + 1}) appx.')
ax.set_ylim(-1.1, 1.1)
ax.set_xlim(-1.5 * np.pi, 1.5 * np.pi)
ax.legend(bbox_to_anchor=(1.02, 1), loc=2, borderaxespad=0.)
fig.subplots_adjust(right=0.75)
```
