```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('data/Pokemon.csv')
pokemon.head()
```

```python
sns.distplot(pokemon['Attack Point']);
```

```python
hist = sns.distplot(
    pokemon['Health Point'], color='black', bins=20, kde=False)
hist.set_title('Health Point Distribution')
hist.set_xlabel('Health Points')
hist.set_ylabel('Frequency');
```

```python
kde_plot = sns.distplot(pokemon['Health Point'], hist=False, rug=True)
kde_plot.set_title('Health Point Density')
kde_plot.set_xlabel('Health Points')
kde_plot.set_ylabel('Unit Probability');
```

```python
sns.set(color_codes=True)
sns.kdeplot(pokemon['Defense Point'], shade=True, color='g');
```

```python
hist = plt.subplots(figsize=(20, 10))
sns.kdeplot(pokemon['Defense Point'])
sns.kdeplot(pokemon['Defense Point'], bw=0.1, label='bs: 0.1');
sns.kdeplot(pokemon['Defense Point'], bw=0.5, label='bs: 0.5');
sns.kdeplot(pokemon['Defense Point'], bw=1, label='bs: 1');
sns.kdeplot(pokemon['Defense Point'], bw=5, label='bs: 5');
sns.kdeplot(pokemon['Defense Point'], bw=10, label='bs: 10');
sns.kdeplot(pokemon['Defense Point'], bw=50, label='bs: 50');
sns.kdeplot(pokemon['Defense Point'], bw=100, label='bs: 100');
plt.xlim([0, 150]);
```

```python
sns.jointplot(x='Attack Point', y='Defense Point', data=pokemon);
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='Attack Point', 
                  y='Defense Point', 
                  data=pokemon, 
                  kind='hex', 
                  color='b');
```

```python
with sns.axes_style('white'):
    sns.jointplot(x='Attack Point', 
                  y='Defense Point', 
                  data=pokemon, 
                  kind='kde', 
                  color='b');
```

```python
sns.kdeplot(pokemon['Attack Point'], 
            pokemon['Defense Point'], 
            clip=(0, 130),);
```

```python
cormat = pokemon.corr()
plt.subplots(figsize=(7, 7))
sns.heatmap(cormat,
            vmax=0.8,
            square=True,
            annot=True,
            fmt='.2f',
            cmap='Blues');
```

# Linear Relationships

```python
sns.lmplot(
    x='Speed Point', y='Defense Point', data=pokemon, size=5, aspect=2);
```

```python
sns.lmplot(x='Attack Point', 
           y='Defense Point', 
           data=pokemon, 
           col='Type 1', 
           hue='Type 2');
```

```python
sns.regplot(x='Attack Point', y='Defense Point', data=pokemon);
```

```python
sns.jointplot(x='Attack Point', 
              y='Defense Point', 
              data=pokemon, 
              xlim=[0, 100],
              ylim=[0, 70],
              kind='reg');
```

```python
sns.pairplot(pokemon,
             size=5,
             aspect=0.9,
             x_vars=['Attack Point', 'Speed Point', 'Defense Point'],
             y_vars=['Total Point'],
             kind='reg');
```

# Categorical and Multipanel

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set(style='whitegrid', color_codes=True)
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=6);
```

```python
sns.swarmplot(x='day', y='tip', data=tips);
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='size');
```

```python
sns.boxplot(x='day', y='tip', data=tips);
```

```python
sns.violinplot(x='day', y='tip', data=tips);
```

```python
sns.violinplot(x='day', y='tip', data=tips, bw=0.1, scale='count');
```

```python
sns.violinplot(x='day', y='tip', data=tips, inner='stick', palette='Set3');
```

```python
sns.barplot(x='day', y='tip', data=tips, hue='size');
```

```python
sns.countplot(x='tip', data=tips, palette='Blues_d')
plt.xticks(rotation=90);
```

```python
f, ax = plt.subplots(figsize=(15, 5))
sns.pointplot(x='day', y='tip', data=tips);
```

```python
f, ax = plt.subplots(figsize=(15, 5))
sns.pointplot(x='day', y='tip', data=tips, hue='size');
```

```python
sns.boxplot(data=tips, orient='h');
```

```python
sns.factorplot(
    x='day', y='tip', data=tips, hue='size', size=5, aspect=2, kind='bar');
```

```python
sns.factorplot(
    x='day', y='tip', data=tips, hue='size', size=5, aspect=2, kind='box');
```

```python
sns.factorplot(x='day', 
               y='tip', 
               data=tips, 
               hue='size', 
               size=5, 
               aspect=2, 
               kind='swarm');
```
