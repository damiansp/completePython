```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
pokemon = pd.read_csv('data/Pokemon.csv')
pokemon.head()
```

# Themes and Figure Styles

```python
plt.subplots(figsize=[10, 5])
sns.distplot(pokemon['Attack Point']);
```

```python
sns.set_style('whitegrid')
plt.subplots(figsize=[10, 5])
sns.distplot(pokemon['Attack Point']);
```

```python
sns.set_style('darkgrid')
plt.subplots(figsize=[10, 5])
sns.distplot(pokemon['Attack Point']);
```

```python
sns.set_style('dark')
plt.subplots(figsize=[10, 5])
sns.distplot(pokemon['Attack Point']);
```

```python
sns.set_style('ticks')
plt.subplots(figsize=[10, 5])
sns.distplot(pokemon['Attack Point']);
```

# Color Palettes

```python
sns.set_palette('ocean_r')
plt.subplots(figsize=(10, 5))
sns.distplot(pokemon['Attack Point']);
```

```python
sns.set_style('whitegrid')
sns.set_palette('Reds')
plt.subplots(figsize=(10, 5))
sns.distplot(pokemon['Attack Point']);
```
